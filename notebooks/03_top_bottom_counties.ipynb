# 03 — Top & Bottom Counties by Prop 13 Benefit

Bar charts ranking counties by:
1. Assessment ratio (lowest = greatest Prop 13 benefit)
2. Annual tax gap per household (dollars underpaid vs. market-rate assessment)

```python
from pathlib import Path
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

ROOT = Path().resolve().parent
df = pd.read_csv(ROOT / 'data' / 'processed' / 'county_prop13.csv')
df = df.dropna(subset=['assessment_ratio', 'tax_gap_per_household'])
df['assessment_ratio_pct'] = df['assessment_ratio'] * 100
```

```python
# --- Bar chart 1: Assessment ratio, all counties ---
ranked = df.sort_values('assessment_ratio')

fig1 = px.bar(
    ranked,
    x='county',
    y='assessment_ratio_pct',
    color='assessment_ratio_pct',
    color_continuous_scale='RdYlGn',
    range_color=[30, 100],
    title='Assessment Ratio by CA County — Lowest First<br>'
          '<sup>Counties with low ratios have large gaps between taxable value and market value</sup>',
    labels={
        'county': 'County',
        'assessment_ratio_pct': 'Assessment Ratio (%)',
    },
    hover_data={'assessment_ratio_pct': ':.1f'},
)
fig1.add_hline(
    y=100, line_dash='dash', line_color='gray',
    annotation_text='100% = fully assessed at market value',
    annotation_position='top right',
)
fig1.update_layout(
    xaxis_tickangle=-60,
    showlegend=False,
    coloraxis_showscale=False,
    height=500,
    margin={'t': 80, 'b': 160},
)
fig1.show()
```

```python
# --- Bar chart 2: Top 20 counties by tax gap per household ---
top20 = df.nlargest(20, 'tax_gap_per_household').sort_values('tax_gap_per_household', ascending=True)

fig2 = px.bar(
    top20,
    y='county',
    x='tax_gap_per_household',
    orientation='h',
    color='tax_gap_per_household',
    color_continuous_scale='Reds',
    title='Top 20 Counties: Annual Tax Gap per Owner-Occupied Household<br>'
          '<sup>Estimated dollars underpaid relative to full market-value assessment</sup>',
    labels={
        'county': '',
        'tax_gap_per_household': 'Annual Tax Gap per Household ($)',
    },
    text='tax_gap_per_household',
)
fig2.update_traces(texttemplate='$%{x:,.0f}', textposition='outside')
fig2.update_layout(
    showlegend=False,
    coloraxis_showscale=False,
    height=520,
    margin={'l': 120, 'r': 80, 't': 80, 'b': 40},
    xaxis=dict(tickprefix='$', tickformat=',.0f'),
)
fig2.show()
```

```python
# --- Bar chart 3: Top 20 by total annual tax gap (millions) ---
top20_total = df.nlargest(20, 'tax_gap_annual_millions').sort_values('tax_gap_annual_millions', ascending=True)

fig3 = px.bar(
    top20_total,
    y='county',
    x='tax_gap_annual_millions',
    orientation='h',
    color='tax_gap_annual_millions',
    color_continuous_scale='Oranges',
    title='Top 20 Counties: Total Annual Property Tax Gap ($M)<br>'
          '<sup>Total estimated revenue gap from Prop 13 assessment limits</sup>',
    labels={
        'county': '',
        'tax_gap_annual_millions': 'Annual Tax Gap ($M)',
    },
    text='tax_gap_annual_millions',
)
fig3.update_traces(texttemplate='$%{x:,.0f}M', textposition='outside')
fig3.update_layout(
    showlegend=False,
    coloraxis_showscale=False,
    height=520,
    margin={'l': 160, 'r': 100, 't': 80, 'b': 40},
    xaxis=dict(tickprefix='$', ticksuffix='M', tickformat=',.0f'),
)
fig3.show()
```
